```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.array([1,2,3,4,5])
plt.figure(figsize=(8,5))
plt.plot(x,x+2)
plt.plot(x,x)
plt.xlabel('x_axis')
plt.ylabel('y_axis')
plt.title('plot for x against y')

plt.show()
```

```python
x= np.linspace(0,10,100)
x
```

```python
plt.plot(x,x,color = 'red',linewidth = 3,linestyle = '-.',label ='linear')
plt.plot(x,x**2,label='squared')
plt.plot(x,x**3,linestyle = '--',label='cubed')
plt.legend()    # to show labels
plt.show()
```

```python
fig,axis = plt.subplots(1,2)       # 1 row and 2 columns     
axis[0].plot(x,x)
axis[1].plot(x,x**2)

plt.show()
```

```python
fig,axis = plt.subplots(2,2)       # 2 row and 2 columns     
axis[0][0].plot(x,x)
axis[1][1].plot(x,x**2)
axis[1,1].set_yscale('log')
plt.show()
```
